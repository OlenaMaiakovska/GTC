```python
import numpy as np
from __future__ import division
from IPython.core.debugger import Tracer; debug_here = Tracer()
from matplotlib import pyplot as plt
```

```python
path='./data.txt'
data = np.loadtxt(path, delimiter="\t")
data=data[:,1:]

#random 5000 is test 
ind=np.random.permutation(len(data))
train=data[ind[:20000],:]
test=data[ind[20000:],:]
x_test=test[:,0]
y_test=test[:,1]

m=np.mean(test,axis=0)
m_x_test=m[0]
m_y_test=m[1]

n_list=range(100,20000,100)
log_l_train=[]
log_l_test=[]
a_list=[]
for n in n_list:
    data=train[0:n,:]
    x=data[:,0]
    y=data[:,1]
    
    m=np.mean(data,axis=0)
    m_x=m[0]
    m_y=m[1]
    
    a=(y-m_y).dot(x.T)/(x-m_x).dot(x.T)
    b=m_y-a*m_x
    a_list.append(a)
    sigma_x=(1/n)*(x-m_x).dot((x-m_x).T)
    diff=y-a*x-b
    diff_test=y_test-a*x_test-b
    
    sigma_y=(1/n)*diff.dot(diff.T) 
    
    log_l_train.append(((-1/(2*sigma_y**2))*(diff.dot(diff.T))+(-1/(2*sigma_x**2))*((x-m_x).dot((x-m_x).T)))/n)
    log_l_test.append((-1/(2*sigma_y**2)*(diff_test.dot(diff_test.T))+(-1/(2*sigma_x**2))*((x_test-m_x_test).dot((x_test-m_x_test).T)))/5000)

# plt.plot(log_l_train)
# plt.plot(log_l_test)
plt.plot(a_list)

plt.show()
```
