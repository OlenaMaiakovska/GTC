#### Part 1. Simple single compartment model

* Physical basic of ion movements
    * Diffusion and Fick's law:    
        $J_{A,diff}=-D_A \frac{d[A]}{dx}$
        
    * Electrical drift (collision with other molecules during movement along electric field):  
        $ J_{A, drift}=-\frac{D_A F}{RT}z_A[A]\frac{dV}{dx} $
    *	Nernst Plank Equation. 
    
    $J_{A}=J_{A,diff}+J_{A,drift}=-D_A \lgroup\frac{d[A]}{dx}+\frac{F}{RT} z_A  [A] \frac{dV}{dx}\rgroup$
* Longitudinal current ???
* Resting membrane potential 
    
    For membrane permeable only to K$^+$:
    
    $K^+$ diffuses $ \Rightarrow $ Electrical charge $ \Rightarrow $ Electrical field that counteracts the diffusion
    
    So the resting membrane potential, it is the potential where Diffusion movement is equal to electrical movement:  $V_A=\frac{RT}{z_AF}\ln\lgroup\frac{[A]_{out}}{[A]_{in}}\rgroup $
* Goldman-Hodgkin-Katz
    1. Constant membrane potantial and idepencence of different ion types (where do we use it to derive GHK current equation?)
    
     $$ j_A=P_Az_AF\frac{z_AFV}{RT}\left ( \frac{[A]_{in}-[A]_{out}\cdot \exp{\frac{-z_A FV}{RT}}}{1- \exp{\frac{-z_A FV}{RT}}}\right) $$
    2. Sum of all $j_A$ to zero to get GHK voltage equation
    
     $$E_m=\frac{RT}{F}\ln \left (\frac{P_{K^+}[K^+]_{out}+P_{Na^+}[Na^+]_{out}+P_{Cl^-}[Cl^-]_{in}}{P_{K^+}[K^+]_{in}+P_{Na^+}[Na^+]_{in}+P_{Cl^-}[Cl^-]_{out}}\right)$$

* Equivalent approximation to linear curcuit: 
    $$j_A=g_A\cdot (^V-E_A)$$
    
* For several ions: 
$$E_m=\frac{ g_{A_{Na^+}} E_{Na^+}+g_{A_{K^+}}E_{K^+}}{g_{A_{Na^+}}+g_{A_{K^+}}}$$

$$R_m=\frac{1}{g_{A_{Na^+}}+g_{A_{K^+}}}$$

#### Part 2. Cable theory

* Assumption of single compartment is not justified for neurons with **thin processes** or **fast
changes of potential**, or for cable-like **long structures**.

* To derive **cable equation** we can use multicompartment model [link] with length of compartment $\delta x \rightarrow 0$
$$V_j=V(x,t), V_{j-1}=V(x-\delta x,t), V_{j+1}=V(x+ \delta x, t)$$

$$\frac{V(x+\delta x)-V(x)}{\delta^2 x}+\frac{V(x-\delta x)-V(x)}{\delta^2 x}=\frac{1}{\delta x}\left (\frac{V(x+\delta x)-V(x)}{\delta x}-\frac{V(x)-V(x-\delta x)}{\delta x}\right )\rightarrow \frac{\partial^2 V(x,t)}{\partial x^2}$$

* Standtrtized form : 
 $$\lambda^2 \frac{\partial^2 \nu (x,t)}{\partial x^2}=\tau_m \frac{\partial \nu (x,t)}{\partial t}+\nu (x,t)-r_m i_e (x,t)$$ 
 Here: 
 1. $\nu (x,t)=V(x,t)-E_m$  
 2. $\lambda^2=\frac{r_m}{r_a}$ - determines the shape of exponential decay along the length of the cable
 3. $\tau_m=c_m r_m$ - time constant
 4. $c_m, r_m, r_a, i_e$ length-normalized parameters


* Steady state solution 
 * The solution that we'll get if we inject $i_e=i_e (x)$ constant current
 $$\lambda^2 \frac{\partial^2 \nu (x,t)}{\partial x^2}=\nu (x,t)-r_m i_e (x,t)$$
 * It is DEQ with solution (using two boundaries conditions )
   * For infinite cable:
 $$\nu (x)=R_{\infty}I_0e^{-\frac{x}{\lambda}}$$
         
   * For finite cable: 
   $$\nu (x)=R_{\infty}I_0\frac{\frac{R_L}{R_{\infty}}\cosh (L-X)+\sinh (L-X)}{\frac{R_L}{R_{\infty}}\sinh (L)+\cosh (L)}$$  
Here: 
        1. $R_{\infty}=\sqrt{r_m r_a}$ 
        2. $X=\frac{x}{\lambda}$
        3. $L=\frac{l}{\lambda}$
 * Boundary conditions
     * **Killed end**: $R_L=0$
     * **Sealed end**: $R_L=\infty$
     * **Leaky end**: $0 < R_L < \infty$


* Time-dependent solution
 * With reparametrization :
$$ \frac{\partial^2 \nu (X,T)}{\partial X^2}=\frac{\partial \nu (X,T)}{\partial T}+\nu (X,T)-r_m i_e (X,T)$$ 
 * Green's function $h(X,T)$ is the solution for a delta pulse $ \delta (X) \delta (T)$ 
 * The general solution is the **superposition** of the responses to different delta pulses

**Lypunov theorem**
If in region $R$ exists positive definite state function $E(x)$ ($ E(x_0)=0, E(x)\geq 0   \forall x \in R$),than
  * If $ \frac{dE(x(t))}{dt} \geq 0 \Rightarrow x_0 $ is stable
  * If $ \frac{dE(x(t))}{dt} > 0 \Rightarrow x_0 $ is asymptotically stable
  * If $ \frac{dE(x(t))}{dt} <0 \Rightarrow x_0 $ is unstable
