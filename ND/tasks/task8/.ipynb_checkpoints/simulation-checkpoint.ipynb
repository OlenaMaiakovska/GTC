```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi=3.1415926
def gabor(x,k0,a,b):
    return (a/((pi**(1/2))*b))*np.exp(-x**2/4*b**2)*np.cos(k0*x)

def signal(x, d):
    return np.exp(-x**2/4*d**2)*(1/(2*pi**(1/2)*d))

def integral(f,A,B,M):
    delta=(B-A)/M
    return (f(np.linspace(A,B,M))*delta).sum()
    
    
    
```

```python
x=np.linspace(-10,10,1000)
plt.plot(x,gabor(x, 3, 1,1))
plt.show()
```

```python
plt.plot(x,signal(x,1))
plt.show()
```

```python
f=lambda x: x
integral(f, 0,2,1000)
```

```python
# params 
tau=10
deltat=10**(-4)
t_start=0
t_end=1
T=int((t_end-t_start)/deltat)
t=np.linspace(t_start,t_end,T)
a=1
b=0.6
d=2
k0=4
A=-10
B=10
M=500
x=np.linspace(A,B,M)
w=lambda x: gabor(x,k0,a,b)
s=lambda x: signal(x,d)

conv=lambda x: integral
```

```python
u=np.zeros((M,T))
W=np.zeros((M,M))
S=np.zeros((M,T))

for i,x_i in enumerate(x):
    S[i,:]=s(x_i)

for i,x_i in enumerate(x):
    for j,x_j in enumerate(x):
        W[i,j]=w(x_i-x_j)
```

```python
plt.plot(S[:,0])
plt.show()
```

```python
for i in range(T-1):
    u[:,i+1]=u[:,i]+(deltat/tau)*(-u[:,i]+W.dot(u[:,i]*deltat)+S[:,i])
```

```python
plt.scatter(x,t,u)
plt.show()
```
